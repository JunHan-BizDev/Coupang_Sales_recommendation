```python
import os
import pandas as pd

path_dir = './coupang/'
dir_list = os.listdir(path_dir)

for d in dir_list:
    base = path_dir + d + '/'
    file_list = os.listdir(base)
    
    # py 
    for f in file_list:
        if '.py' in f:
            file_list.remove(f)
            

    
    #  
    data = pd.read_csv(base + file_list[0], error_bad_lines=False)
    
    for file in file_list[1:9]:
        temp = pd.read_csv(base + file, error_bad_lines=False)
        data = pd.concat([data,temp])

        #  (1~120    )
        data = data.reset_index()
        data = data.drop(['index'],axis=1)
    
    # result.csv  
    del file_list[len(file_list)-1]
    
    #  
    data2 = pd.read_csv(base + file_list[9], error_bad_lines=False)
    
    k = 10
    
    for file in file_list[10:]:
        if file == 'result.csv':
            continue
        print(base, file, k)
        
        temp = pd.read_csv(base + file, error_bad_lines=False)
        data2 = pd.concat([data2,temp])
        
        #  (1~120    )
        data2 = data2.reset_index()
        data2 = data2.drop(['index'],axis=1)
        k+=1
        
    #result    
    product_id = file[14:file.find('csv')-1]
    df_temp = data2[data2['reg_date'] > '2020.06.01'] # 2020 6 1  
    df_temp = df_temp.groupby('product_id').agg(['count'])
    df_temp = df_temp.ratings
    data = pd.merge(left = data, right = df_temp, how = 'left', on = 'product_id')
    
    
    #  
    data = data.drop_duplicates(['product_id'])
        
    
    #    csv 
    data.to_csv(base + 'result.csv')   
    data2.to_csv(base + 'result_review.csv')
```
